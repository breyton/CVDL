用 pandas 讀取 csv 文件後可以發現前 30 個 column 是坐標值，最後一個 column 是圖片像素值的字串 (以空白鍵區隔)，接下來我們就以下的函數來做資料的讀取以及前處理，請盡量仔細理解這個函數

```python
# 過濾有缺失值的 row
print(data.shape)
data = data.dropna()
print(data.shape)
print( type(data['Image']))
print( type(data['Image'][0]))
# 將圖片像素值讀取由Series變為 numpy array 的形態
data['Image'] = data['Image'].apply(lambda img: np.fromstring(img, sep=' ')).values 
print( type(data['Image']))
print( type(data['Image'][0]))
```

```python
# 單獨把圖像 array 抽取出來
imgs = np.vstack(data['Image'].values)/255
print( type(imgs))
```

```python
imgs.shape
```

```python
(np.vstack(data['Image'])/255)[:,0]
```

```python
data.columns[-1:]
```

```python
# reshape 為 96 x 96
imgs = imgs.reshape(data.shape[0], 96, 96)
# 轉換為 float
imgs = imgs.astype(np.float32)

# 提取坐標的部分
print(imgs.shape)
points = data[data.columns[:-1]].values

# 轉換為 float
points = points.astype(np.float32)

# normalize 坐標值到 [-0.5, 0.5]
points = points/96 - 0.5
```

```python
print("圖像資料:", imgs.shape, "\n關鍵點資料:", points.shape)
```

```python
plt.imshow(imgs[0], cmap='gray')
```

```python
sample_img = imgs[0]
sample_points = points[0]
```

```python
markers=sample_points.reshape(15,2)
```

```python
x,y=markers[0]*96+48
```
